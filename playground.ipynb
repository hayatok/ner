```python
import spacy
nlp = spacy.load('ja_ginza_electra')
doc = nlp('銀座でランチをご一緒しましょう。')
for sent in doc.sents:
    for token in sent:
        print(token.i,
            token.orth_,
            token.lemma_,
            token.norm_,
            token.morph.get("Reading"),
            token.pos_,
            token.morph.get("Inflection"),
            token.tag_,
            token.dep_,
            token.head.i,
        )
    print('EOS')
```

```python
import spacy
from spacy import displacy
nlp = spacy.load('ja_ginza_electra')
```

```python
#ここでいろんな処理が行われる
doc = nlp("2019年現在、日本の航空会社では日本航空、全日本空輸、AIRDOの3社で使用されている。") 

#固有表現抽出の結果の描画
displacy.render(doc, style="ent", jupyter=True)
```
